Although there have been lot of studies undertaken in the past on factors affecting life expectancy considering demographic variables, income composition and mortality rates. It was found that affect of immunization and human development index was not taken into account in the past. This study will focus on immunization factors, mortality factors, economic factors, social factors and other health related factors as well. Since the observations this dataset are based on different countries, it will be easier for a country to determine the predicting factor which is contributing to lower value of life expectancy. This will help in suggesting a country which area should be given importance in order to efficiently improve the life expectancy of its population.

* Country: Country(193 countries)
* Year: This project include data from year 2000-2015
* Status: Developed or Developing
* Life expectancy:  Average number of years a person is expected to live.
* Adult Mortality: Adult Mortality Rates of both sexes (probability of dying between 15 and 60 years per 1000 population)
* infant deaths: Number of Infant Deaths per 1000 population
* Alcohol: Alcohol, recorded per capita (15+) consumption (in litres of pure alcohol)
* percentage expenditure: Expenditure on health as a percentage of Gross Domestic Product per capita(%). Shows how much of the country's economic resources are allocated to health.
* Hepatitis B: Hepatitis B (HepB) immunization coverage among 1-year-olds (%)
* Measles: number of reported cases per 1000 population
* BMI: Average Body Mass Index of entire population
* under-five deaths: Number of under-five deaths per 1000 population
* Polio: Polio (Pol3) immunization coverage among 1-year-olds (%)
* Total expenditure: General government expenditure on health as a percentage of total government expenditure (%). Highlights the priority given to healthcare within government budgets.
* Diphtheria: Diphtheria tetanus toxoid and pertussis (DTP3) immunization coverage among 1-year-olds (%)
* HIV/AIDS: Deaths per 1000 live births HIV/AIDS (0-4 years)
* GDP: Gross Domestic Product per capita (in USD). An indicator of a country’s economic health and its correlation with life expectancy.
* Population: Population of the country
* thinness 1-19 years: Prevalence of thinness among children and adolescents for Age 10 to 19 (% ). Indicates malnutrition in older children and adolescents.
* thinness 5-9 years: Prevalence of thinness among children for Age 5 to 9(%). Reflects malnutrition in younger children.
* Income composition of resources: Human Development Index in terms of income composition of resources (index ranging from 0 to 1). Measures access to resources that contribute to human development, like education and income.
* Schooling: Average number of years of schooling in the country.

* Mean/Median: Replace null values with the mean or median of the column. This works well for numerical data that is approximately normally distributed (mean) or skewed (median).
* Mode: Replace with the most frequent category in the column.

```python
# columns with approximately normal distribution
nor_col = ['Schooling','Income composition of resources','Total expenditure','BMI','Life expectancy']

for col in nor_col:
    df[col] = df[col].fillna(df[col].mean())
```

```python
from category_encoders import TargetEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=0)
```

```python
x_train = pd.DataFrame(x_train, columns = x.columns)
x_test = pd.DataFrame(x_test, columns = x.columns)
```

```python
# Target encoding for 'Country'
target_encoder = TargetEncoder(cols=['Country'])
x_train['Country'] = target_encoder.fit_transform(x_train['Country'], y_train)
x_test['Country'] = target_encoder.transform(x_test['Country'])
```

```python
# One-hot encode 'Status'
x_train = pd.get_dummies(x_train, columns=['Status'], drop_first=True)
x_test = pd.get_dummies(x_test, columns=['Status'], drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

x_train['Year'] = le.fit_transform(x_train['Year'])
x_test['Year'] = le.fit_transform(x_test['Year'])
```

```python
x_train.head()
```

```python
x_train.columns
```

```python
numerical_cols = ['Country', 'Year', 'Adult Mortality', 'infant deaths', 'Alcohol',
       'percentage expenditure', 'Hepatitis B', 'Measles', 'BMI', 'Polio',
       'Total expenditure', 'Diphtheria', 'HIV/AIDS', 'GDP', 'Population',
       'thinness 10-19 years', 'thinness 5-9 years',
       'Income composition of resources', 'Schooling']
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train[numerical_cols])  # Fit and transform the training data
x_test = scaler.transform(x_test[numerical_cols])   
```

```python
# Train the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Perform cross-validation after training
cv_scores = cross_val_score(model, x_train, y_train, cv=5, scoring='r2')
```

```python
print("Cross-Validation r2:", cv_scores)
print("Mean r2:", cv_scores.mean())
```

```python
y_train_pred = model.predict(x_train)
train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
print("Test RMSE:", train_rmse)
r2_train = r2_score(y_train, y_train_pred)
print('R2 Score:', r2_train)
```

```python
# Test the model
y_test_pred = model.predict(x_test)
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
print("Test RMSE:", test_rmse)
r2_cv = r2_score(y_test, y_test_pred)
print('R2 Score:', r2_cv)
```

```python
def predict_life_expectancy(input_data, target_encoder, scaler, model, numerical_cols):
    # Convert input data to DataFrame
    input_df = pd.DataFrame([input_data])
    
    # Target encode the 'Country' column
    if 'Country' in input_df.columns:
        input_df['Country'] = target_encoder.transform(input_df['Country'])
    
    # One-hot encode 'Status' column
    if 'Status' in input_df.columns:
        input_df = pd.get_dummies(input_df, columns=['Status'], drop_first=True)

    if 'Year' in input_df.columns:
        input_df['Year'] = le.transform(input_df['Year'])
    
    # Ensure input_df has the same columns as training data
    input_df = input_df.reindex(columns=numerical_cols, fill_value=0)
    
    # Standardize numerical columns
    input_df = scaler.transform(input_df)
    
    # Predict life expectancy
    prediction = model.predict(input_df)
    return prediction[0]
```

```python
# Example usage of the function
random_data = {
    'Country': 'Afghanistan',
    'Year': 2008,
    'Status': 'Developing',
    'Adult_Mortality': 287,
    'infant_deaths': 80,
    'Alcohol': 0.03,
    'percentage_expenditure': 25.873925,
    'Hepatitis_B': 64.0,
    'Measles': 1599,
    'BMI': 15.7,
    'under-five_deaths': 110,
    'Polio': 64.0,
    'Total_expenditure': 8.33,
    'Diphtheria': 64.0,
    'HIV/AIDS': 0.1,
    'GDP': 373.361116,
    'Population': 2729431.0,
    'thinness_1-19_years': 18.8,
    'thinness_5-9_years': 18.9,
    'Income_composition_of_resources': 0.433,
    'Schooling': 8.7
}

# Predict life expectancy for the example data
predicted_life_expectancy = predict_life_expectancy(
    random_data, target_encoder, scaler, model, numerical_cols
)
print("Predicted Life Expectancy:", predicted_life_expectancy)
```
